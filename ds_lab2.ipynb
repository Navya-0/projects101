```python
a = pd.DataFrame(oh.fit_transform(df[["parental level of education","race/ethnicity"]]).toarray())
df = df.join(a)
df
```

outliers detection

```python
df.boxplot(["math score", "reading score", "writing score"])
```

```python
df.min()
```

```python
q1=df["math score"].quantile(0.25)
q3=df["math score"].quantile(0.75)
iqr= q3-q1
iqr
```

```python
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
```

outliers values


```python
df[df['math score'] > upper_limit]
df[df['math score'] < lower_limit]
```

trimming outliers, here we only used lower limits because there were no outliers beyond upper limit

```python
new_df = df[df["math score"] > lower_limit]
new_df
```
